```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('class.csv')
df.head()
```

```python
pd.value_counts(df["like_social"])
```

```python
pd.crosstab(df["gender"], df["like_social"])
```

```python
crosstab= pd.crosstab(df["like_social"],df["gender"], normalize="columns")
crosstab
```

```python

sns.countplot(x="gender", hue="like_social", data=df)
plt.show()
```

For students in the MDDD program, females generally like social media more than males. It is also important to note that there are more females (18) than males (6) so a more accurate depiction would show the percentage of females who agree with this statement.

# Exploring categorical & numerical variables

The distributions below show the number of phones owned by gender.

```python
sns.catplot(x="gender", y="phones_owned", kind="violin", data=df)
plt.title('Number of phones owned by gender')
plt.show()
```

```python
sns.countplot(x="gender", hue="phones_owned", data=df)
plt.show()
```

The distributions below shows the number of phones owned by the degree to which the individuals like social media.

```python
sns.countplot(x="like_social", hue="phones_owned", data=df)
plt.show()
```

We can see that contrary to common belief, those who do not like social media still have purchase around 6-8 phones.

Below we take a look at the relationship between those who like social media and the number of phones owned, by gender.

```python
sns.catplot(x="like_social", y="phones_owned", hue="gender", kind="violin", data=df)
plt.title('Number of phones owned by gender')
plt.show()
```

On average, the degree to which users like social media does not really have any effect on the number of phones owned for either gender.
