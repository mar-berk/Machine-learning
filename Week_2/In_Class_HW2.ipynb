# Correlation

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt #Useful to include matplotlib when plotting with Seaborn, because Seaborn is built on it.
```

```python
df = pd.read_csv('funda-amsterdam.csv')
df.head()
```

Create a subset of the dataframe to only show the variables of interest- area, rooms, and price.

```python
df_subset = df[['area', 'rooms', 'price']] #making a new dataframe with a subset
sns.pairplot(df_subset)
plt.show()
```

We need to remove the faulty entries because they are improperly skewing the dataset. 

# Visualizing the dataframe with outliers removed

It is hard to tell what is a natural otlier of the dataset and what is faulty data entry so we will create some common sense limits. 
Each entry must have:

* Price above 30,000
* Area above 10 m2
* Rooms above 0

```python
df = df[(df['price'] > 30000) & (df['area'] > 10) & (df['rooms'] > 0)] #we want to subset the original dataset, otherwise we lose data
df_subset = df[['area', 'rooms', 'price']]

sns.pairplot(df_subset)
plt.show()
```

Now we will make a heatmap showing this data.

First we must correlate the variables

```python
corr = df[['price', 'area', 'rooms']].corr() 
corr
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap="YlGnBu") #the last two arguments set the column names
plt.show()
```

# Price by number of rooms

```python
sns.lmplot(x='rooms', y='price', data=df_subset, scatter_kws={'alpha':0.1}) #fit_reg=False leaves out the regression line; see below
plt.xlabel('Number of rooms') #use matplotlib to refine Seaborn plot
plt.ylabel('Price')
plt.title('Price by number of rooms')
plt.show()
```

As shown by the graph, generally the price goes up as the number of rooms increases.

```python
from sklearn.linear_model import LinearRegression #we need this specific model from sk-learn

x = df_subset[['rooms']] #to use sk-learn, we need to create lists of the two variables
y = df_subset['price']

lm = LinearRegression() #this creates a new LR model
lm.fit(x, y) #this "fits" the model

b0 = lm.intercept_ #lm.coef_ gets a list of coefficients (excluding the intercept). [0] gets the actual number from the list
b1 = lm.coef_[0] #gets the intercept

print(f"The regression line is equal to y = {b0:.2f} + {b1:.2f}X") #.0f formats to 2 decimals.
```

```python
from scipy.stats.stats import pearsonr #We will need the correlation function from the scientific package scipy
print(pearsonr(df_subset['rooms'],df_subset['price']))
```
