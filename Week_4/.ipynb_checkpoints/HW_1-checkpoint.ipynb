# In-class Assignment 1

Predicting bol.com sales. 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('bol_sales.csv')
df = df.dropna()
df.head(10)
```

# Exploratory data analysis

```python
df.corr().sort_values('spent_bol', ascending=False)
```

```python
sns.pairplot(df)
plt.show()
```

# Variable Transformations

```python
df['income_log'] = df[('income')].apply(math.log10)
df.head()
```

```python
y = df['spent_bol'] #We need to take out the money spent log as our Y-variable
X = df[['distance_center','website_visits', 'income_log']] #our x variable
X.index = df['id'] #this stores the id in the column names, so we don't lose track of them later
X.head()
```

Now that we have selected our variables let's make a training set and a test set

# Building the Model

```python
#This built-in function from sk-learn splits the data set randomly into a train set and a test set
#By stating random_state = 1, we use one particular "random state" (we could use any number, it's a so-called "random seed"). 
#This means if we run the code again, it will produce the same results. Which can be handy.
#test_size = 0.3, so I'm splitting the data into 70% training data and 30% test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
X_train.head() #The train data
```

Now let's train the linear model

```python
lm = LinearRegression() #create the model
model = lm.fit(X_train, y_train) #train the model
```

Now let's look at the coefficients

```python
coef = pd.DataFrame(X.columns.values, lm.coef_) #this creates a df with a column with the column names, and the coefficients of the model
coef
```

```python
y_test_p = lm.predict(X_test) #Generate predictions for the test data. So we use X.test, not X.train
r2_score(y_test, y_test_p)
```

The R^2 is approximately 0.88, which means 880% of the variation is explained by the model. That's maybe too high. Let's calculate the RMSE: the root mean squared error.

```python
math.sqrt(mean_squared_error(y_test, y_test_p))
```

The RMSE is 105.37. Which means that we are typically about 105.37 off in our predicted rating (which goes from 0-10). This is great news. Finally, let's plot the predicted value against the actual rating.

```python
sns.scatterplot(y_test, y_test_p)
plt.xlabel('Actual money spent log')
plt.ylabel('Predicted money spent')
plt.plot([500, 1700], [500, 1700], color='red', lw=3) #Plot a diagonal length

plt.show()
```

Here we see the predicted shares log (y axis) and the actual shares log (x axis). Any values along the red line have been predicted accurately. As we can see the model is relatively accurate,
