# Classification with _Random Forest_

```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #We need this to split the data
```

```python
df = pd.read_csv('creditcard.csv')
#df = df[['Survived','Pclass', 'Age', 'SibSp', 'Sex', 'Parch']] #make this selection before you dropna
df = df.dropna() #first get rid of rows with empty cells 
df.head(30)
```

```python
df['Class'].value_counts()
```

```python
from sklearn.preprocessing import normalize #get the function needed to normalize our data.

X = df.loc[:,'Time' : 'Amount'] #create the X matrix

y = df['Class'] #create the y-variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
#the order of the above variables is very important
X_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier() #create a Random forest-classifier 
rf = rf.fit(X_train, y_train) #this fits the k-nearest neigbor model with the train data
rf.score(X_test, y_test) #calculate the fit on the test data
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = rf.predict(X_test) #the predicted values
cm = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['Not fraud (actual)', 'Fraud (actual)'], columns = ['Not fraud (predicted)', 'Fraud (predicted)']) 
conf_matrix
```

The way to read this is that of the fraud, 51 are correctly predicted as 'Fraud', and 4 that were predicted as fraud were actually not fraud. And of those who were predicted as not fraud, 29930 were predicted correctly, while 15 actually were fradu. The _recall_ and _precision_ for the fraud predictions:

$recall = \frac{51}{ 15 + 51} = .77$

$precision = \frac{51 }{ 4 + 51} = .92$



The model is pretty good at predicting what is not fraud. It did detect 4 items as fraud when it was not actually fraud but for a bank it is better to be precise than to be accurate, because they can call customers to verify purchases so it is better to be safe.

# Try with new Parameters

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50) #create a Random forest-classifier 
rf = rf.fit(X_train, y_train) #this fits the k-nearest neigbor model with the train data
rf.score(X_test, y_test) #calculate the fit on the test data
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = rf.predict(X_test) #the predicted values
cm = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['Not fraud (actual)', 'Fraud (actual)'], columns = ['Not fraud (predicted)', 'Fraud (predicted)']) 
conf_matrix
```

The way to read this is that of the fraud, 50 are correctly predicted as 'Fraud', and 0 that were predicted as fraud were actually not fraud. And of those that were predicted as not fraud, 29934 were predicted correctly, while 16 actually were fraud. The _recall_ and _precision_ for the fraud predictions:

$recall = \frac{50}{ 16 + 50} = .75$

$precision = \frac{50 }{ 0 + 50} = 1$



When using the new number of estimators (n_estimators=50 as opposed to the default of 100) we got an increase in _precision_- meaning all of the prediced fraud was actually fraud, this may seem good, but for a bank this is not the relationship we are most focused on.
With the new number of estimators, we got a decrease in _recall_ - meaning less of the real fraud was detected.

For a bank, recall is the most important relationship- we want to be able to detect all of the actual fraud. So in this case using the default number of estimators was better than n_estimators= 50
