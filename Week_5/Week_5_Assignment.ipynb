The data set _voice.csv_ was created to identify a voice as male or female, based upon acoustic properties of the voice and speech. The dataset consists of 3,168 recorded voice samples, collected from male and female speakers. The voice samples are pre-processed by acoustic analysis in R using the seewave and tuneR packages, with an analyzed frequency range of 0hz-280hz More information on this dataset can be found here: https://www.kaggle.com/primaryobjects/voicegender/home

The dataset is cleaned, split into a training and test set and the K-nearest neighbor algorithm is used to predict which recordings are from female speakers.

```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #We need this to split the data
```

```python
df = pd.read_csv('voice.csv')
df.head(30) #let's take a look at the dataset
```

Now that we can see the dataset, let's take a look at how the variables correlate with eachother so we can select the variables we want to include in our prediction model.

```python
df.corr()
```

Unfortunately the variable we need: _label_ does not show up in this correlation table so we need to make dummie variables for the _label_ variable.

# Data Cleaning & Variable Selection

```python
dummies = pd.get_dummies(df['label'])#make dummie variables for the gender
dummies
df = pd.concat([df, dummies], axis=1) #the axis=1 means: add it to the columns
df.head(10)
```

Now there are two new variables- female and male. Let's take a look at the correlation matrix now.

```python
df.corr()
```

The variables that correlate with our new dummy variables, male and female, the most are used for the model. 

```python
df = df[['IQR','sp.ent', 'meanfun', 'sd', 'Q25','centroid','sfm','meanfreq', 'female']] #make this selection before you dropna
df = df.dropna() #first get rid of rows with empty cells 
df['female'].value_counts()
```

# Building the Model

For this prediction, we are using the _K nearest neighbor_ algorithm. 

KNN is an algorithm that determines the value of a data point by the data points around it (neighbors). KNN classifier determines the class of a data point by majority voting principle. Basically this means that if k is set to 3, the classes of 3 closest points are checked. Prediction is done according to the majority class. 

```python
from sklearn.preprocessing import normalize #get the function needed to normalize our data.

X = df[['IQR','sp.ent', 'meanfun', 'sd', 'Q25','centroid','sfm','meanfreq']] #create the X matrix
#X = normalize(X) #normalize the matrix to put everything on the same scale

y = df['female'] #create the y-variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
#the order of the above variables is very important
X_train.head()
```

Now we will use the _KNeighborsClassifier_ from sklearn.

```python
from sklearn.neighbors import KNeighborsClassifier #the object class we need

knn = KNeighborsClassifier(n_neighbors=3) #create a KNN-classifier with 3 neighbors
knn = knn.fit(X_train, y_train) #this fits the k-nearest neigbor model with the train data
```

# Model Evaluation

Now that we have cleaned the dataset, split the dataset into a training and a test set and preformed the KNN algorithm with 3 neighbors. Let's evaluate the results, beginning with accuracy.

```python
knn.score(X_test, y_test) #calculate the fit on the test data
```

The accuracy is 97% this is very good but does not necessarily tell the whole story of our predictions. Let's take a look at the confusion matrix to get a clearer picture.

```python
from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test) #the predicted values
cm = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['Not female (actual)', 'Female (actual)'], columns = ['Not female (predicted)', 'Female (predicted)']) 
conf_matrix
```

The way to read this is that of the predicted female voices, 446 are correctly predicted as _female_, 12 voices that were predicted as female were actually male. And of those who were predicted as _not female_, 482 were predicted correctly, while 11 predicted _not female_ voices were actually female voices. The _recall_ and _precision_ for the voice predictions:

$recall = \frac{446}{ 11 + 446} = .98$

$precision = \frac{446 }{ 446 + 12} = .97$

We might improve our scores by trying out different values of _k_.

The recall is 98% which means that 98% of the female voices where correctly predicted to be female voices.

The precision is 97% which means that 97% of the predicted female voices were actually female voices.
