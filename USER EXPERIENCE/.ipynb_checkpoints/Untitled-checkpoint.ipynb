# Survey Results

```python
# importing for drawing plots
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# open the dataframe
df = pd.read_csv("Online Shopping Survey.csv")
df.head()
```

```python
sns.countplot(x="How important are a companies moral/ ethical values to you?", hue="Does your perception of Amazon.com's ethics/ moral values effect how frequently you shop there?", data=df)
plt.show()
```

```python
sns.countplot(x="When shopping online, which of the following is MOST important to you?", data=df)
plt.show()
```
